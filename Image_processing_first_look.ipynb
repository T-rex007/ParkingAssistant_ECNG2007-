```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
VID_PATH = "data/parking_lot.mp4"
```

```python
cap0 = cv2.VideoCapture(VID_PATH)
while(cap0.isOpened()):
    ret, frame = cap0.read()
    img = frame
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 200)
    cv2.imshow("some", edges)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap0.release()
```

```python
import numpy as np
if __name__ == '__main__':
    r = cv2.selectROI(img)
    imCrop = img[int(r[1]):int(r[1]+r[3]), int(r[0]):int(r[0]+r[2])]
    cv2.imshow("Image", imCrop)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(VID_PATH)
#fourcc = cv2.VideoWriter_fourcc(*'MPEG')
#out = cv2.VideoWriter('jumpFenceOut2.avi',fourcc, 30.0,(imCrop.shape[0],imCrop.shape[1]))
while(1):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
    edges = cv2.Canny(gray, 50, 200)
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold = 10, maxLineGap = 1 )
    for line in lines:
        x1, y1, x2, y2 = line[0]
        
        
        cv2.line(frame, (x1,y1), (x2,y2), (255, 0, 0), 3)
    cv2.imshow("cam", frame)
    #print("here")
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
